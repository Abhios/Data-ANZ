## PREDICTION

```python
salary['gender'].replace({'F' : 1, 'M' : 0}, inplace= True)
```

```python
salary=pd.get_dummies(salary)
salary
```

```python

from sklearn.model_selection import train_test_split as split
```

```python
X=salary.drop(['amount','txn_description_PAY/SALARY'],axis=1)
y=salary[['amount']]
```

```python
X.shape,y.shape
```

```python
x_train,x_test,y_train,y_test=split(X,y,test_size=0.3,random_state=5)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

Lin_reg=LinearRegression()

# fit the data
Lin_reg.fit(x_train,y_train)

# predict (training samples)
y_pred=Lin_reg.predict(x_train)
print("\nTraining score : ")
print("Mean squared error: %.2f"% mean_squared_error(y_train, y_pred))
print('R2 score: %.2f' % r2_score(y_train, y_pred))

# predict (training samples)
y_pred=Lin_reg.predict(x_test)
print("\nTesting score : ")
print("Mean squared error: %.2f"% mean_squared_error(y_test, y_pred))
print('R2 score: %.2f' % r2_score(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score,mean_absolute_error
from sklearn.ensemble import RandomForestRegressor

rfc = RandomForestRegressor(bootstrap=True, criterion='mse', max_depth=13,
                      max_features='sqrt',n_estimators=100,
                      n_jobs=None, oob_score=False)   
rfc.fit(x_train,y_train)

# predict (training samples)
y_pred = rfc.predict(x_train)

print("\nTraining score : ")
print("Mean squared error: %.2f"% mean_squared_error(y_train, y_pred))
print('R2 score: %.2f' % r2_score(y_train, y_pred))


y_pred=rfc.predict(x_test)
print("\nTesting score : ")
print("Mean squared error: %.2f"% mean_squared_error(y_test, y_pred))
print('R2 score: %.2f' % r2_score(y_test, y_pred))
```
